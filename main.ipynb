```python
# fazer a importacao da base de dados utilizando o pandas
import pandas as pd 

tabela = pd.read_csv('cancelamentos.csv')
tabela = tabela.drop(columns='CustomerID') # descartar coluna que nao acrescenta na análise de dados
tabela = tabela.dropna() # descartar linhas que possuem valores vazios
display(tabela)
```

```python
# analisando a porcentagem de pessoas que cancelaram 

display(tabela['cancelou'].value_counts())
display(tabela['cancelou'].value_counts(normalize = True))
```

```python
# analise de cada aspecto para o cancelamento:
# a partir dos 50 anos, todas as pessoas dessa faixa cancelaram
# as mulheres em média cancelam mais do que os homens
# o tempo omo cliente e a frequencia de uso não importa no cancelamento
# quanto maior o numero de ligacoes ao callcenter e o número de dias de atraso, maior será o índice de cancelamento
# o plano de assinatura não influencia
# o contrato mensal possui taxa de cancelamento altíssima
# a partir dos 500 reais, o público tende a cancelar menos
# se a pessoa demora a usar, tende a cancelar
```

```python
# alterações para diminuir o número de cancelamentos

tabela = tabela[tabela['duracao_contrato']!='Monthly']
tabela = tabela[tabela['total_gasto']>500]
tabela = tabela[tabela['ligacoes_callcenter']<5]
display(tabela['cancelou'].value_counts(normalize=True))

# com essas três modificações, o número de cancelamento reduz drasticamente em relação ao seu valor anterior
```
